```python
from matplotlib import pyplot as plt
from pytransform3d.transformations import plot_transform
from pytransform3d.plot_utils import make_3d_axis

ax = make_3d_axis(ax_s=1, unit='m', n_ticks=6)
plot_transform(ax=ax)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pytransform3d.rotations as pr
a = np.array([0.6, 0.3, 0.9])
b = np.array([0.4, 0.8, 0.2])
ax = pr.plot_basis()
ax.view_init(azim=20, elev=20)
ax.set_xlim((0 ,1))
ax.set_ylim([0, 1])
ax.set_zlim((0, 1))
plt.tight_layout()
plt.show()
```

```python
ax = pr.plot_basis()
pr.plot_bivector(ax=ax, a=a, b=b)
plt.tight_layout()
plt.show()
```

```python
from pytransform3d.plot_utils import plot_vector
plot_vector(
    start=np.array([-0.3, -0.2, -0.3]),
    direction=np.array([1.0, 1.0, 1.0]),
    s=0.5,
    ax_s=0.5,
    lw=0,
    color="green"
)
plt.show()
```

```python
from pytransform3d.plot_utils import plot_sphere, remove_frame
from pytransform3d.transformations import plot_transform
ax = plot_sphere(radius=0.5, wireframe=True, alpha=0.9, color='k', ax_s=0.5)
plot_transform(ax=ax, A2B = np.eye(4), s=0.3, lw=3)
remove_frame(ax)
plt.show()
```

```python
import pytransform3d.transformations as pt
mean = pt.transform_from(R=np.eye(3), p = np.array([0, 0, 0.5]))
cov = np.diag([0.001, 0.001, 0.001, 0.001, 0.001, 0.001])
rng = np.random.default_rng(0)
ax = None
poses = np.array([
    pt.random_transform(rng=rng, mean= mean, cov=cov) for _ in range(10)])
for pose in poses:
    ax = pt.plot_transform(ax=ax, A2B=pose, s=0.3)
remove_frame(ax)
plt.show()
```
